```python
import pandas as pd
import matplotlib.pyplot as plt

equities = ["TCS","RELIANCE", "INFY", "HDFC", "HDFCBANK","BAJFINANCE","LT","ZEEL"]
#equities = ["LT"]
path='C:/Personal/Finance/Stocks/Trading/Historical_Data/New/nse-listed-1384-companies-data/HISTORICAL_DATA/HISTORICAL_DATA/'

for equity in equities:
    filename = path + equity +"_data.csv"
    df = pd.read_csv(filename)
    df = df.drop(['high', 'low', 'volume'], axis = 1)
    df['Date'] = pd.to_datetime(df.Date)
    df['Year'] = df['Date'].dt.strftime('%Y')
    Years = ["2012","2013","2014","2015","2016","2017","2018"]
    #Years = ["2014"]
    Returns = []
    
    for year in Years:
        df_temp=df[df['Year'] == year]
        df_temp['Profit'] = df_temp['open'] - df_temp['close'].shift()
        df_temp['Profit'] = df_temp['Profit'].fillna(0)
        Returns.append(df_temp['Profit'].sum())
        #print(df_temp)
        
    #print(Returns)
    chart_name=equity + " BTST Performance"    
    plt.style.use('ggplot')
    x_pos = [i for i, _ in enumerate(Years)]
    plt.bar(x_pos, Returns,color='g')
    plt.xlabel("")
    plt.ylabel("Returns")
    plt.title(chart_name)
    plt.rcParams["figure.figsize"] = (19,5)

    plt.xticks(x_pos,Years)
    filename="c:\\Personal\\" + equity + "_BTST.png"
    plt.savefig(filename)
    #plt.show()
```
