```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("C:\\Personal\\Finance\\Stocks\\Trading\\Data\CONTRACT_01022020.csv")

def clean(x):
    x = x.replace(":", "")[0:4]
    return x


df = data[data['Ticker'] == 'ACC20FEBFUT.NFO']


datetime_series = pd.to_datetime(df['Time'])
datetime_index = pd.DatetimeIndex(datetime_series.values)

df3=df.set_index(datetime_index)


ohlc_dict = {                                                                                                             
'Open':'first',                                                                                                    
'High':'max',                                                                                                       
'Low':'min',                                                                                                        
'Close': 'last',                                                                                                    
'Volume': 'sum'
}

df3 = df3.resample('15T', closed='left', label='left').apply(ohlc_dict)

#data['index'].strftime('%h%m , % r') 

#data.index.astype('str')

df3['Time'] = df3.index

type(df3.index)

```
