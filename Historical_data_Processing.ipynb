```python
import pandas as pd
import os
import json
from datetime import date
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook
import matplotlib.image as image
fig = plt.figure()

path = 'C:/Personal/Finance/Stocks/Trading/Historical_Data/2020/nse/'

Files = []
for r, d, f in os.walk(path):
    for file in f:
        if ".csv" in file:
            #print(file)
            Files.append(os.path.join(r, file))

symbols = []
profits = []
            
def analyse_data(filename):    
    df = pd.read_csv(filename)
    df['volIncr'] = df['Total Traded Quantity'].gt(df['Total Traded Quantity'].shift())
    #df=df[df['volIncr'] == True]
    no_of_stocks=1
    df['Profit'] = no_of_stocks*(df['Open Price'] - df['Prev Close'])
    #df['Profit'].plot()
    
    #profits.append(df['Profit'].sum().astype(str))
    profits.append(df['Profit'].sum().astype(float))
    symbols.append(str(df.iloc[0]['Symbol']))
    #print(str(df.iloc[0]['Symbol']) + ":" + df['Profit'].sum().astype(str))
    
    

#print(Files)          
for file in Files:
    analyse_data(file)
    
plt.style.use('ggplot')
x_pos = [i for i, _ in enumerate(symbols)]
plt.bar(x_pos, profits)
plt.xlabel("")
plt.ylabel("Returns")
plt.title("BTST - 2020")
plt.rcParams["figure.figsize"] = (19,5)

plt.xticks(x_pos,symbols)

plt.show()
```
