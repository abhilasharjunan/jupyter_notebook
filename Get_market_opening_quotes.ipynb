```python
from datetime import date
from datetime import datetime
import calendar
import requests
from nsetools import Nse
nse=Nse()
from bs4 import BeautifulSoup
from pprint import pprint
from requests_html import HTMLSession
import json
import csv
import shutil
import pandas as pd


index_entries = {
        "NIFTY AUTO" : "cnxAuto",
        "NIFTY BANK" : "bankNifty",
        "NIFTY IT" : "cnxit",
        "NIFTY FIN SERVICE" : "cnxFinance", 
        "NIFTY FMCG" : "cnxFMCG",
        "NIFTY MEDIA" : "cnxMedia",
        "NIFTY METAL" : "cnxMetal",
        "NIFTY PHARMA" : "cnxPharma",
        "NIFTY PSU BANK" : "cnxPSU",
        "NIFTY PVT BANK" : "niftyPvtBank",
        "NIFTY REALTY" : "cnxRealty"
    }


index_list = ["NIFTY PVT BANK","NIFTY PSU BANK", "NIFTY AUTO","NIFTY BANK","NIFTY FIN SERVICE","NIFTY FMCG","NIFTY IT",
              "NIFTY MEDIA","NIFTY METAL","NIFTY PHARMA","NIFTY REALTY"]

pct_index={}


def get_data_from_nse(url):
    session = HTMLSession()
    resp = session.get(url)
    soup = BeautifulSoup(resp.html.html, 'html.parser')#soup = BeautifulSoup(page.content, 'html.parser')

    data = json.loads(str(soup))
    return data


def get_top_stocks(Gainers_or_Losers):
    Base_url="https://www1.nseindia.com/live_market/dynaContent/live_analysis/gainers/niftyGainers1.json"
    if(Gainers_or_Losers == "L"):
        Base_url = "https://www1.nseindia.com/live_market/dynaContent/live_analysis/losers/niftyLosers1.json"
    data = get_data_from_nse(Base_url)
    #table = pd.DataFrame.from_dict(data["data"])
    #drop_col = ['lastCorpAnnouncement','lastCorpAnnouncementDate','series','netPrice','turnoverInLakhs','tradedQuantity']
    #table.drop(drop_col, axis = 1,inplace=True)
    #table['Open_High'] = table['highPrice'].equals(table['openPrice'])
    #table['Open_Low'] = table['lowPrice'].equals(table['openPrice'])
    #print(table['symbol'].head(5))
    
    for i in range(0,min(5,len(data['data']))):
        print(data['data'][i]['symbol'].replace('&amp;','&'))

def get_open_high_low():
    Base_url="https://www1.nseindia.com/live_market/dynaContent/live_watch/stock_watch/niftyStockWatch.json"
    data = get_data_from_nse(Base_url)
    #table = pd.DataFrame.from_dict(data["data"])
    #drop_col = ['lastCorpAnnouncement','lastCorpAnnouncementDate','series','netPrice','turnoverInLakhs','tradedQuantity']
    #table.drop(drop_col, axis = 1,inplace=True)
    #table['Open_High'] = table['highPrice'].equals(table['openPrice'])
    #table['Open_Low'] = table['lowPrice'].equals(table['openPrice'])
    stocks_values=data['data']
    #print(stocks_values[0])
    open_low=[]
    open_high=[]
    for stock in stocks_values:
        if (stock['open'] == stock['low']):
            open_low.append(stock['symbol'])
        if (stock['open'] == stock['high']):
            open_high.append(stock['symbol'])
    
    return open_low,open_high


def sort_values_dict(entries, rev=False):
    sorted_orders=sorted(entries.items(),key = lambda x:x[1], reverse=rev)
    return sorted_orders

def get_indexes():
    url="https://www1.nseindia.com/live_market/dynaContent/live_watch/stock_watch/liveIndexWatchData.json"
    data = get_data_from_nse(url)
    for i in range(0,55):
    #for i in [ 2, 3, 10, 13, 17, 23, 32, 34, 36, 49, 50]:
        if data['data'][i]['indexName'] in index_list:
            index_name = data['data'][i]['indexName']
            pct_change = data['data'][i]['percChange']
            pct_index[index_name] = float(pct_change)
            #print(data['data'][i])
    return sort_values_dict(pct_index, True)

def get_top_stocks_from_index(index, reverse = True):
    stock_index={}
    entry = index_entries[index]
    url="https://www1.nseindia.com/live_market/dynaContent/live_watch/stock_watch/" + entry + "StockWatch.json"
    data = get_data_from_nse(url)
    for i in range(0, len(data['data'])):
        value = float(data['data'][i]['ltP'].replace(",",""))
        if value > 100:
            stock_name = data['data'][i]['symbol']
            pct_change = data['data'][i]['per']
            stock_index[stock_name] = float(pct_change)
            #print(data['data'][i])
    return sort_values_dict(stock_index,reverse)


def get_pre_open_data():
    stock_index={}
    count=5
    url="https://www.nseindia.com/api/market-data-pre-open?key=NIFTY"
    data = get_data_from_nse(url)['data']
    print("============================================")
    print("Pre-open Top stocks at "+ str(datetime.now()))
    print("============================================")
    print("Symbol | Price | percentChange | Traded Vol       ")
    print("============================================")
    for i in data:
        if float(i['metadata']['lastPrice']) > 100:
        #print(i)
            stock_name = i['metadata']['symbol']
            price = str(i['metadata']['lastPrice'])
            percentChange = str(i['metadata']['pChange'])[0:5]
            traded_vol = str(i['detail']['preOpenMarket']['totalTradedVolume'])
            print(stock_name + " | " + price + " | " + percentChange + " | "+ traded_vol)
            count -= 1
            if count < 1:
                break
                
                
get_pre_open_data()

print("============================================")    
print(datetime.now())
print("\n==============\nTop Gainers\n==============")
get_top_stocks("G")
print("\n==============\nTop Loosers\n==============")
get_top_stocks("L")

open_low,open_high = get_open_high_low()
print("\n==============\nOpen = Low Stocks\n==============")

for stock in open_low:
    print(stock)
    
print("\n==============\nOpen = High Stocks\n==============")
for stock in open_high:
    print(stock)

indexes = get_indexes()

print("=======================================================")
top_index_gainer = indexes[0]
print("Top Index Gainer : " + top_index_gainer[0] + " Change% : " + str(top_index_gainer[1]) )

top_gainer_in_index = get_top_stocks_from_index(top_index_gainer[0])[0]
print("Top Gainer : " + top_gainer_in_index[0] + " Change% : " + str(top_gainer_in_index[1]))

print("========================================================")

top_index_loser = indexes[-1]
print("Top Index Loser : " + top_index_loser[0] + " Change% : " + str(top_index_loser[1]) )

top_loser_in_index = get_top_stocks_from_index(top_index_loser[0],False)[0]
print("Top Loser : " + top_loser_in_index[0] + " Change% : " + str(top_loser_in_index[1]))
print("=========================================================")

#print(get_indexes()[-1])
#print(get_top_stocks_from_index(get_indexes()[-1][0],False)[0])
```
